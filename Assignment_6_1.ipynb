## 5. How  do  you count the frequency  of each  value  appearing in the  given array  of integers?

```python
import numpy as np
import pandas as pd

data = np.array([0, 5, 4, 0, 4, 4, 3, 0, 0, 5, 2, 1, 1, 9])

# Convert the NumPy array to a pandas Series
data_series = pd.Series(data)

# Count the frequency of each value
frequency_counts = data_series.value_counts()

frequency_counts.sort_index(inplace=True)

# Print the frequency of each value
print("Frequency of each value:")
print(frequency_counts)
```

## 11. Create a random array of 3 rows and 3 columns and sort it according to the first, second, or third column.

```python
import numpy as np

# Create a random array of 3 rows and 3 columns
random_array = np.random.rand(3, 3)
print("Original 3x3 Random Array:")
print(random_array)

# Sort by the first column (index 0)
sorted_by_first_col = random_array[random_array[:, 0].argsort()]
print("\nSorted by the first column:")
print(sorted_by_first_col)

# Sort by the second column (index 1)
sorted_by_second_col = random_array[random_array[:, 1].argsort()]
print("\nSorted by the second column:")
print(sorted_by_second_col)

# Sort by the third column (index 2)
sorted_by_third_col = random_array[random_array[:, 2].argsort()]
print("\nSorted by the third column:")
print(sorted_by_third_col)
```

## 12. Create a four dimensions array to get the sum over the last two axes at once.

```python
import numpy as np

# Create a four-dimensional NumPy array
# Example: a 2x3x4x5 array
four_d_array = np.arange(2 * 3 * 4 * 5).reshape(2, 3, 4, 5)
print("Original 4D array:")
print(four_d_array)

# Get the sum over the last two axes (axis 2 and axis 3)
sum_over_last_two_axes = np.sum(four_d_array, axis=(-2, -1))

# Print the result
print("\nSum over the last two axes:")
print(sum_over_last_two_axes)
```

## 13. Create a random array and swap two rows of an array.

```python
import numpy as np

# Create a random array (e.g., 5x4)
random_array = np.random.rand(5, 4)
print("Original array:")
print(random_array)

# Swap two rows (e.g., swap row 1 and row 3)
# Using advanced indexing with a list of indices
swapped_array = random_array[[0, 3, 2, 1, 4], :]

print("\nArray with rows swapped:")
print(swapped_array)

# Alternatively, you can swap in-place
array_to_swap_inplace = np.random.rand(5, 4)
print("\nOriginal array for in-place swap:")
print(array_to_swap_inplace)

array_to_swap_inplace[[1, 3]] = array_to_swap_inplace[[3, 1]]
print("\nArray after in-place swap of row 1 and 3:")
print(array_to_swap_inplace)
```

## 14. Create a random matrix and compute the matrix rank.

```python
import numpy as np

# Create a random matrix (e.g., 4x5)
random_matrix = np.random.rand(4, 5)
print("Original random matrix:")
print(random_matrix)

# Compute the matrix rank
matrix_rank = np.linalg.matrix_rank(random_matrix)

# Print the matrix rank
print("\nMatrix rank:", matrix_rank)
```

## Data collection

Load the `middle_tn_schools.csv` file into a pandas DataFrame and display its descriptive statistics.


```python
df_schools = pd.read_csv('/content/sample_data/middle_tn_schools.csv')
print("First 5 rows of the DataFrame:")
display(df_schools.head())
print("\nDescriptive statistics of the DataFrame:")
display(df_schools.describe())
```

## Data grouping

Select the `reduced_lunch` and `school_rating` columns and group the data by `school_rating`. Also check the distribution of `stu_teach_ratio` column.


```python
# Select 'reduced_lunch' and 'school_rating' columns
selected_columns = df_schools[['reduced_lunch', 'school_rating']]

# Group the selected data by 'school_rating'
grouped_by_rating = selected_columns.groupby('school_rating')

# Display the groups (optional, can show a few groups or info)
print("Grouping the data by 'school_rating':")
for name, group in grouped_by_rating:
    print(f"\nSchool Rating: {name}")
    display(group.head()) # Display first few rows of each group
    if name == 1 or name == 5: # Display only a few groups to keep the output manageable
      break
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a histogram of the stu_teach_ratio
plt.figure(figsize=(10, 6))
sns.histplot(data=df_schools, x='stu_teach_ratio', kde=True)

# Add title and labels
plt.title('Distribution of Student-Teacher Ratio')
plt.xlabel('Student-Teacher Ratio')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

## Correlation analysis

Calculate and display the correlation between `school_rating` and `reduced_lunch` and also `stu_teach_ratio`.


```python
# Calculate the correlation matrix for the selected columns
correlation_matrix = selected_columns.corr()

# Extract the correlation between school_rating and reduced_lunch
correlation_school_rating_reduced_lunch = correlation_matrix.loc['school_rating', 'reduced_lunch']

# Print the correlation coefficient
print(f"The correlation between school_rating and reduced_lunch is: {correlation_school_rating_reduced_lunch}")
```

```python
# Calculate the correlation between stu_teach_ratio and school_rating
correlation_stu_teach_rating = df_schools['stu_teach_ratio'].corr(df_schools['school_rating'])

# Print the correlation coefficient
print(f"The correlation between student-teacher ratio and school_rating is: {correlation_stu_teach_rating}")
```

## Scatter plot

Create a scatter plot to visualize the relationship between `school_rating` and `reduced_lunch`.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=selected_columns, x='reduced_lunch', y='school_rating')

# Add title and labels
plt.title('Relationship between School Rating and Percentage of Students on Reduced Lunch')
plt.xlabel('Percentage of Students on Reduced Lunch')
plt.ylabel('School Rating')

# Display the plot
plt.show()
```

## Correlation matrix

Compute and visualize the correlation matrix for relevant columns in the DataFrame.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant numerical columns
relevant_columns = [
    'school_rating', 'size', 'reduced_lunch', 'state_percentile_16',
    'state_percentile_15', 'stu_teach_ratio', 'avg_score_15', 'avg_score_16',
    'full_time_teachers', 'percent_black', 'percent_white', 'percent_asian',
    'percent_hispanic'
]

# Calculate the correlation matrix
correlation_matrix = df_schools[relevant_columns].corr()

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of School Variables')
plt.show()
```
